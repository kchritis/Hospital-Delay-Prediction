```python
# 1. Φέρνουμε τα απαραίτητα εργαλεία
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score

# 2. Φτιάχνουμε "ψεύτικα" δεδομένα για να δουλέψει ο κώδικας (αφού δεν έχουμε το αρχείο σου ακόμα)
data = {
    'Age': np.random.randint(18, 80, 100),
    'Hour': np.random.randint(8, 16, 100),
    'DayOfWeek': np.random.randint(0, 5, 100),
    'Delay': np.random.choice([0, 1], 100, p=[0.7, 0.3]) # 0=Στην ώρα του, 1=Καθυστέρηση
}
df = pd.DataFrame(data)

# 3. Εκπαιδεύουμε τον αλγόριθμο Random Forest
X = df[['Age', 'Hour', 'DayOfWeek']] # Τι κοιτάζει ο αλγόριθμος
y = df['Delay'] # Τι προσπαθεί να προβλέψει
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)

# 4. Βγάζουμε το αποτέλεσμα (Accuracy)
predictions = model.predict(X_test)
print(f"Ακρίβεια Μοντέλου: {accuracy_score(y_test, predictions)*100:.2f}%")

# 5. Φτιάχνουμε το Heatmap (Confusion Matrix)
plt.figure(figsize=(6,4))
sns.heatmap(confusion_matrix(y_test, predictions), annot=True, cmap='Blues')
plt.title('Heatmap Προβλέψεων')
plt.show()
```

```python
Σε αυτό το στάδιο βλέπουμε την αξιολόγηση του μοντέλου Random Forest. Η ακρίβεια 70% δείχνει ότι το μοντέλο μπορεί να προβλέψει με μεγάλη επιτυχία τις καθυστερήσεις.
```

```python
# Δείχνει ποιος παράγοντας είναι ο πιο σημαντικός
features = ['Age', 'Hour', 'DayOfWeek']
importances = model.feature_importances_

plt.figure(figsize=(8,4))
sns.barplot(x=importances, y=features, palette='viridis')
plt.title('Ποιοι παράγοντες προκαλούν την καθυστέρηση;')
plt.show()
```
